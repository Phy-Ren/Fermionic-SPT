Three Majorana fermions inserted: gamma_B(34) -> gm[0], gamma_B(24) -> gm[1], gamma_B(23) -> gm[2]
The gm is a List of 0,1 denoting the absence or presence of the inserted Majorana fermions.
The gm is related to n_1, n_2, omega_2, s_1:
gm[0] = dn_2(1234), gm[1] = dn_2(0134), gm[2] = dn_2(0123), gm[3] = dn_2(0234), gm[4] = dn_2(0124)

```python
gamma_insert = (19, 19, 13, 19, 17)
gamma_total = (19, 17, 13)


def Connection(gm):
    #gm = Z2 number of gamma_insert
    #gm_total = Z2 number of gamma_total
    global gm_total
    gm_total = [gm[0]^gm[1]^gm[3], gm[4], gm[2]]
    gamma_insert_pair = []
    for i in range(len(gm_total)):
        for j in range(i+1,len(gm_total)):
            if gm_total[i]*gm_total[j] == 1:
                gamma_insert_pair = [gamma_total[i], gamma_total[j]]
    
    if len(gamma_insert_pair) == 0:
        #print("No Nontrivial Pairing")
        return [], []
    
    insert_pairs_list = []
    pairs_list = []
    needed_pairs_set = set()

    for pair in totalSet:
        if pair[1] == gamma_insert_pair[0]:
            pairs_list.append((0, 1, pair[1], pair[2]))
            if pair[2] == gamma_insert_pair[1]:
                needed_pairs_set.add((0, 1, pair[1], pair[2]))
                return needed_pairs_set, insert_pairs_list

        elif pair[2] == gamma_insert_pair[0]:
            pairs_list.append((0, 2, pair[1], pair[2]))
            if pair[1] == gamma_insert_pair[1]:
                needed_pairs_set.add((0, 1, pair[1], pair[2]))
                return needed_pairs_set, insert_pairs_list

    insert_pairs_list.append(deepcopy(pairs_list))
    #checked_pair = (index of start gamma, left or right, left gamma, right gamma)
    #The left or right part is where the gamma start, and we should always check the opposite one, final gamma.
    #If checked_pair[1] == 1 we check checked_pair[3]; if checked_pair[1] == 2 we check checked_pair[2].
    for i in range(0,10):
        pairs_list = []
        for j in range(len(insert_pairs_list[i])):
            for pair in totalSet:
                #Find all pairs contain final gamma
                if pair[1] == insert_pairs_list[i][j][2] and pair[2] == insert_pairs_list[i][j][3]:
                    continue

                elif pair[1] == insert_pairs_list[i][j][3] and insert_pairs_list[i][j][1] == 1:
                    pairs_list.append((j, 1, pair[1], pair[2]))
                    if pair[2] == gamma_insert_pair[1]:
                        needed_pairs_set.add((j, 1, pair[1], pair[2]))
                        return needed_pairs_set, insert_pairs_list

                elif pair[2] == insert_pairs_list[i][j][3] and insert_pairs_list[i][j][1] == 1:
                    pairs_list.append((j, 2, pair[1], pair[2]))
                    if pair[1] == gamma_insert_pair[1]:
                        needed_pairs_set.add((j, 2, pair[1], pair[2]))
                        return needed_pairs_set, insert_pairs_list

                elif pair[1] == insert_pairs_list[i][j][2] and insert_pairs_list[i][j][1] == 2:
                    pairs_list.append((j, 1, pair[1], pair[2]))
                    if pair[2] == gamma_insert_pair[1]:
                        needed_pairs_set.add((j, 1, pair[1], pair[2]))
                        return needed_pairs_set, insert_pairs_list

                elif pair[2] == insert_pairs_list[i][j][2] and insert_pairs_list[i][j][1] == 2:
                    pairs_list.append((j, 2, pair[1], pair[2]))
                    if pair[1] == gamma_insert_pair[1]:
                        needed_pairs_set.add((j, 2, pair[1], pair[2]))
                        return needed_pairs_set, insert_pairs_list
    
        if len(pairs_list) == 0:
            print("Fail to Find Pairs")
            return [], []
        insert_pairs_list.append(deepcopy(pairs_list))
    print("Fail to Find Pairs")
    return [], []


def NeededPairsSets(gm):
    needed_pairs_set, insert_pairs_list = Connection(gm)
    if len(needed_pairs_set) == 0:
        return []
    needed_pair = needed_pairs_set.pop()
    needed_pairs_set.add((needed_pair[2], needed_pair[3]))
    for i in range(len(insert_pairs_list)-1, -1, -1):
        needed_pair = insert_pairs_list[i][needed_pair[0]]
        needed_pairs_set.add((needed_pair[2], needed_pair[3]))
    return needed_pairs_set
```

```python
def Find_Pairs(pairs, pairSet):
    #The input data is of following types, and we want to find whether pairs are in pairSet.
    #pairs = {(32, 23), (23, 33)}
    #pairSet = {((-0-1j), 10, 13), (-1j, 14, 3)}
    pairs_in_pairSet = []
    for pair_2 in pairs:
        for pair_3 in pairSet:
            if pair_2[0] == pair_3[1] and pair_2[1] == pair_3[2]:
                pairs_in_pairSet.append(pair_3)
    return pairs_in_pairSet



def CommutativeDiagram(gm):
    global needed_pairs_set
    needed_pairs_set = NeededPairsSets(gm)
    sort_pair_list = []
    #phase_list = []
    phase_total = 1
    #sort_pair_list is in the order from left to right in the formula
    pairs_in_pairSet = Find_Pairs(needed_pairs_set, pentagonSet_1)
    for pair in pairs_in_pairSet:
        sort_pair_list.append(pair[1])
        sort_pair_list.append(pair[2])
        phase_total = phase_total*pair[0]
    if gm[0] == 1:
        sort_pair_list.append(gamma_insert[0])

    pairs_in_pairSet = Find_Pairs(needed_pairs_set, pentagonSet_2)
    for pair in pairs_in_pairSet:
        sort_pair_list.append(pair[1])
        sort_pair_list.append(pair[2])
        phase_total = phase_total*pair[0]
    if gm[1] == 1:
        sort_pair_list.append(gamma_insert[1])
    
    pairs_in_pairSet = Find_Pairs(needed_pairs_set, pentagonSet_3)
    for pair in pairs_in_pairSet:
        sort_pair_list.append(pair[1])
        sort_pair_list.append(pair[2])
        phase_total = phase_total*pair[0]
    if gm[2] == 1:
        sort_pair_list.append(gamma_insert[2])

    pairs_in_pairSet = Find_Pairs(needed_pairs_set, pentagonSet_4)
    for pair in pairs_in_pairSet:
        sort_pair_list.append(pair[1])
        sort_pair_list.append(pair[2])
        phase_total = phase_total*pair[0]
    if gm[3] == 1:
        sort_pair_list.append(gamma_insert[3])
    
    pairs_in_pairSet = Find_Pairs(needed_pairs_set, pentagonSet_5)
    for pair in pairs_in_pairSet:
        sort_pair_list.append(pair[1])
        sort_pair_list.append(pair[2])
        phase_total = phase_total*pair[0]
    if gm[4] == 1:
        sort_pair_list.append(gamma_insert[4])

    return sort_pair_list, phase_total
    
```

```python
def BubbleSort(sort_list):
    count = 0
    for i in range(1,len(sort_list)):
        for j in range(len(sort_list)-i):
            if sort_list[j] > sort_list[j+1]:
                sort_list[j], sort_list[j+1] = sort_list[j+1], sort_list[j]
                count = count + 1
    return sort_list, count

```
